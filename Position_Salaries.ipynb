```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/Shivam-kumar20/Position_Salary/main/Position_Salaries.csv'
data = pd.read_csv(url)
```

```python
print(data)  # Actual Dataframe
```

```python
sns.scatterplot(x='Level',y='Salary',data=data,c='red')
# plt.plot(data.Level,data.Salary)
plt.show()
```

### Linear Regression

```python
x=data['Level'].values.reshape(-1,1)
y=data['Salary'].values.reshape(-1,1)

# linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
# Model is loaded

model = lr.fit(x,y)
# Model is Trained

pred = lr.predict(x)
# print(pred)
print(f'The r2Score is {r2_score(y,pred)}')
```

```python
# plotting the graph
plt.scatter(data.Level, data.Salary)
plt.plot(data.Level, pred,c='r') 
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
# plt.grid()
plt.figure(figsize=(8, 6)) # Set the figure size (width, height)
plt.show()
```

```python
Linear_Pred=[]

for i in pred:
    Linear_Pred.extend(i)
Linear_Pred

Lin= pd.DataFrame({'X_Train':data.Level,
                 'Y_Train':data.Salary,
                 'Linear_Pred':Linear_Pred})
Lin
```

### Polynomial Regression

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
poly = PolynomialFeatures(degree = 4)
x_poly = poly.fit_transform(x)
# Model is Loaded

lin2 = lr.fit(x_poly,y)
# Model is trained

pred1 = lr.predict(x_poly)
# print(pred1)

print(f'The r2Score is {r2_score(y,pred1)}')
```

```python
# plotting the graph
plt.scatter(data.Level, data.Salary)
plt.plot(data.Level, pred1,c='r') 
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
# plt.grid()
plt.figure(figsize=(8, 6)) # Set the figure size (width, height)
plt.show()
```

```python
Poly_Pred=[]

for i in pred1:
    Poly_Pred.extend(i)
Poly_Pred

Poly= pd.DataFrame({'X_Train':data.Level,
                 'Y_Train':data.Salary,
                 'Poly_Pred':Poly_Pred})
Poly
```
